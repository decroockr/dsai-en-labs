# Hoofdstuk 3: oefeningen leerpad

```python
print(f"oefening1.a : {1 - (5/6)**4}")
print(f"oefening1.b : {1 - (35/36)**24}")

print(f"oefening2: {1- 1/100 * 2/100 * 2/100}")

print("oefening 3:")
df = pd.DataFrame({
    "x": np.arange(start=2, stop=13),
    "noemer": [1, 2, 3, 4, 5, 6, 5, 4, 3, 2, 1]
})
df['fx'] = df.noemer/36
df['xfx'] = df.x * df.fx
df['xfx_inbreng'] = (df.x - 8) * df.fx
print(f"verificatie kansen: {sum(df.fx)}")
print(f"verwachte waarde: {sum(df.xfx)}") # 3.a
print(f"verwachte waarde met inbreng: {sum(df.xfx_inbreng)}") # 3.b
print("E(X-a) = E(X) - a") # 3.c

print("oefening 4:")
df['variantie'] = (df.x - np.average(df.x))**2 * df.fx
print(f"variantie: {sum(df.variantie)}") # 4.a
df['xfx_helft'] = 1/2 * df.x * df.fx
df['xfx_helft_inbreng'] = (1/2 * df.x - 8) * df.fx
print(f"verwachte waarde met inbreng en halve winst: {sum(df.xfx_helft_inbreng)}") # 4.b
print("E(X/a) = E(X)/a") # 4.c
```
